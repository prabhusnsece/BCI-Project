```python
import mne
import numpy as np
import os
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from tensorflow.keras.layers import *
from tensorflow.keras.models import Model
```

```python
data_path = "/content/BCI_IV_2b/"

files = [
    "B0101T.gdf",
    "B0102T.gdf",
    "B0103T.gdf"
]
```

```python
def extract_trials(file):

    raw = mne.io.read_raw_gdf(file, preload=True, verbose=False)

    # KEEP ONLY EEG CHANNELS (removes EOG)
    raw.pick(picks='eeg')

    raw.set_eeg_reference('average')

    raw.filter(8,30)

    events, event_id = mne.events_from_annotations(raw)

    epochs = mne.Epochs(
        raw,
        events,
        event_id={'769':event_id['769'],'770':event_id['770']},
        tmin=0.5,
        tmax=3,
        baseline=None,
        preload=True,
        event_repeated='merge'
    )

    X = epochs.get_data()
    y = epochs.events[:, -1]

    y = np.where(y == event_id['769'],0,1)

    return X,y
```

```python
X_list=[]
y_list=[]

for f in files:

    full_filepath = os.path.join(data_path, f)

    X,y = extract_trials(full_filepath)

    X_list.append(X)
    y_list.append(y)

X = np.concatenate(X_list)
y = np.concatenate(y_list)

print("Dataset shape:",X.shape)
print("Class distribution:",np.bincount(y))
```

```python
def extract_trials(file):

    raw = mne.io.read_raw_gdf(file, preload=True, verbose=False)

    # keep only EEG channels (removes EOG)
    raw.pick(picks='eeg')

    raw.set_eeg_reference('average')

    raw.filter(8,30)

    # get events
    events, event_id = mne.events_from_annotations(raw)

    print("Event mapping:", event_id)

    # obtain correct IDs
    left_id = event_id['769']
    right_id = event_id['770']

    epochs = mne.Epochs(
        raw,
        events,
        event_id={'left':left_id,'right':right_id},
        tmin=0.5,
        tmax=3,
        baseline=None,
        preload=True,
        event_repeated='merge'
    )

    X = epochs.get_data()
    y = epochs.events[:, -1]

    y = np.where(y == left_id,0,1)

    return X,y
```

```python
X_list = []
y_list = []

for f in files:

    full_filepath = os.path.join(data_path, f)

    X,y = extract_trials(full_filepath)

    X_list.append(X)
    y_list.append(y)

X = np.concatenate(X_list)
y = np.concatenate(y_list)

print("Dataset shape:", X.shape)
print("Class distribution:", np.bincount(y))
```

```python
n_trials, n_channels, n_times = X.shape

X = X.reshape(n_trials, -1)

scaler = StandardScaler()
X = scaler.fit_transform(X)

X = X.reshape(n_trials, n_channels, n_times)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    stratify=y,
    random_state=42
)

print("Train:", X_train.shape)
print("Test:", X_test.shape)
```

```python
X_train = X_train[..., np.newaxis]
X_test = X_test[..., np.newaxis]

print(X_train.shape)
```

```python
def EEGNet():

    input1 = Input(shape=(X_train.shape[1],626,1))

    x = Conv2D(8,(1,64),padding='same',use_bias=False)(input1)
    x = BatchNormalization()(x)

    x = DepthwiseConv2D((X_train.shape[1],1),depth_multiplier=2,use_bias=False)(x)
    x = BatchNormalization()(x)
    x = Activation('elu')(x)

    x = AveragePooling2D((1,4))(x)
    x = Dropout(0.5)(x)

    x = SeparableConv2D(16,(1,16),padding='same',use_bias=False)(x)
    x = BatchNormalization()(x)
    x = Activation('elu')(x)

    x = AveragePooling2D((1,8))(x)
    x = Dropout(0.5)(x)

    x = Flatten()(x)

    output = Dense(2,activation='softmax')(x)

    model = Model(inputs=input1,outputs=output)

    return model
```

```python
model = EEGNet()

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()

model.fit(
    X_train,
    y_train,
    epochs=150,
    batch_size=16,
    validation_split=0.2,
    verbose=1
)
```

```python
loss,acc = model.evaluate(X_test,y_test)

print("Final Accuracy:",acc*100)
```

```python
from tensorflow.keras.layers import GlobalAveragePooling2D
from tensorflow.keras.layers import Reshape, Multiply
```

```python
def channel_attention(input_feature):

    channel = input_feature.shape[-1]

    avg_pool = GlobalAveragePooling2D()(input_feature)
    avg_pool = Reshape((1,1,channel))(avg_pool)

    dense1 = Dense(channel//2, activation='relu')(avg_pool)
    dense2 = Dense(channel, activation='sigmoid')(dense1)

    scale = Multiply()([input_feature, dense2])

    return scale
```

```python
def EEGNet():

    input1 = Input(shape=(6,626,1))

    x = Conv2D(16,(1,64),padding='same',use_bias=False)(input1)
    x = BatchNormalization()(x)

    x = DepthwiseConv2D((6,1),depth_multiplier=2,use_bias=False)(x)
    x = BatchNormalization()(x)
    x = Activation('elu')(x)

    # Channel Attention
    x = channel_attention(x)

    x = AveragePooling2D((1,4))(x)
    x = Dropout(0.25)(x)

    x = SeparableConv2D(16,(1,16),padding='same',use_bias=False)(x)
    x = BatchNormalization()(x)
    x = Activation('elu')(x)

    x = AveragePooling2D((1,8))(x)
    x = Dropout(0.25)(x)

    x = Flatten()(x)

    output = Dense(2,activation='softmax')(x)

    model = Model(inputs=input1,outputs=output)

    return model
```

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stop = EarlyStopping(
    monitor='val_loss',
    patience=30,
    restore_best_weights=True
)

model = EEGNet()

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(
    X_train,
    y_train,
    epochs=300,
    batch_size=16,
    validation_split=0.2,
    callbacks=[early_stop],
    verbose=1
)
```

```python
loss,acc = model.evaluate(X_test,y_test)

print("Final Accuracy:",acc*100)
```
